# Problem
In this homework we have to solve the Cutting Stock Problem using Simulated Annealing. The simulated annealing algorithm is an optimization method which mimics the slow cooling of metals. I will explain my code in this report and some ways i used to enhance it.

# Explanation
In this problem we have to extract the customer's request from the given master rolls. Now if we consider the client request as an array and start from the beginning of the array and extract the request from the remaining master roll until the request is bigger than the master roll. In that case we will throw the extra away and get a new master roll. We will keep doing this until we've created all of the requests. In this method our search space is all of the permutations of the array of client's requests and the neigbor is defined as the swap mutation of the original sample.

# Tempreture Function
Selecting rating of temperature change has a crucial effect on the quality of the solution. If the temperature in the annealing process is not lowered slowly and enough time is not spent at each temperature, the process could get trapped in a local minimum state for the internal energy. The linear function strategies do not guarantee convergence towards the global optimum, but they
converge more rapidly towards a strong minimum (strong minima are solutions that lie in the near vicinity of the global minimum).

```python
import random
import math
import numpy as np

class sample:
  def __init__(self, order, fitness):
    self.order = order
    self.fitness = fitness
  
  def __str__(self):
    return f"{self.fitness}\n{self.order}"
```

```python
class CSP:
  def __init__(self, T, neig_count, function_const, repeat=None):
    self.T = T
    self.repeat = repeat
    self.log_size = 0
    self.neig_count = neig_count
    self.function_const = function_const
    self.best_result = None

  def open_file(self, file):
    f = open(file, "r")
    txt = f.readlines()
    self.log_size = int(txt[0].split()[-1])
    self.request = [float(j) for j in txt[3].split(", ")]

  def neighbors(self, order):
    neig_list = []
    for i in range(self.neig_count):
      order_copy = order[:]
      x = random.randint(0, len(order)-1)
      y = random.randint(0, len(order)-1)
      order_copy[x], order_copy[y] = order_copy[y], order_copy[x]
      neig_list.append(sample(order_copy, self.fitness(order_copy)))
    return neig_list
  
  def fitness(self, order):
    count = 1
    size = self.log_size
    for req in order:
      if req > size:
        count+=1
        size = self.log_size-req 
      else:
        size -= req
    return count
        
  def prob(self, diff, T):
    if random.uniform(0,1) <= np.exp(-(diff/T)):
      return True
    return False
  
  def check_neighbor(self, x, neig_list, T_copy):
    res = neig_list[0] #x
    for neig in neig_list:
      if neig.fitness < res.fitness:
        res = neig
      # elif self.prob(abs(neig.fitness- res.fitness), T_copy) == True:
      #   res = neig
        
    return res
  
  def SA_algorithm(self, file_name):
    self.open_file(file_name)
    random_req = self.request
    random.shuffle(random_req)
    x = sample(random_req, self.fitness(random_req))
    T_copy = self.T
    self.best_result = x
    
    for i in range(self.repeat):
      T_copy = T_copy* self.function_const
      neig_list = self.neighbors(x.order)
      neig = self.check_neighbor(x, neig_list, T_copy)
      if neig.fitness < x.fitness:
        x = neig
      elif self.prob(abs(neig.fitness- x.fitness), T_copy) == True:
        x = neig

      if neig.fitness < self.best_result.fitness:
        self.best_result = neig
    print(self.best_result)
```
